```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import griddata 

sample_g3_fcm = pd.read_csv('Gradient3data.csv')
pro_red = sample_g3_fcm.loc[sample_g3_fcm['red_prochloro'].notnull()] #grouped data into dataframe with no empty slots

plt.title('Red chloropyll of Prochlorococcus')
plt.xlabel('Latitude')
plt.ylabel('Depth (m)')

num_red = len(pro_red['red_prochloro'].to_list())

x_red = pro_red['lat'].to_list()
y_red = pro_red['depth'].to_list()
z_red = np.log(pro_red['red_prochloro'].to_list())

plt.ylim(max(y_red), 0)

# number of samples to generate
numcols, numrows = num_red, num_red

# specify grid limits to color in data
# Creates array of x and y values
xi_red = np.linspace(min(x_red), max(x_red), numcols)
yi_red = np.linspace(min(y_red), max(y_red), numrows)

# Create 2-D rectangular grid from 1-D inputs (arrays)
xi_red, yi_red = np.meshgrid(xi_red, yi_red)

# grid of interpolated data
zi_red = griddata((x_red, y_red), z_red, (xi_red, yi_red), method = 'linear')

# zi = parameter of data; extent = bounding box of coordinates; aspect = affect graph size;
# interpolation = color box in relation to nearest box, origin = placing [0,0] on graph
abundance = plt.imshow(zi_red, 
                       extent=[min(x_red), max(x_red), min(y_red), max(y_red)], 
                       aspect=0.10, 
                       interpolation = 'nearest', 
                       origin = 'lower')

# create color bar
plt.colorbar(abundance, label = 'Red chlorophyll')

plt.scatter(x_red, y_red)
```
