```python
import matplotlib.pyplot as plt
import plotly as ply
import pandas as pd

# creating Prochlorococcus depth and diameter scatter graph
depth_file = pd.read_csv("Depthdata.csv")

# creating labels of the graph
plt.title("Depth and diameter of Prochlorococcus")
plt.xlabel("Diameter (micrometers)")
plt.ylabel("Depth (m)")

plt.ylim(225, 0)

# choosing 2 stations I wan to plot
s1 = depth_file.loc[depth_file["station"] == 1]
s4 = depth_file.loc[depth_file["station"] == 4]

pro_x_axis1 = s1["diameter_prochloro"].to_list()
pro_y_axis1 = s1["depth"].to_list()

pro_x_axis4 = s4["diameter_prochloro"].to_list()
pro_y_axis4 = s4["depth"].to_list()

plt.scatter(pro_x_axis1, pro_y_axis1, label = "Station 1")
plt.scatter(pro_x_axis4, pro_y_axis4, label = "Station 4")

plt.legend(loc = "upper right")
```

```python
# creating Synechococcus depth and diameter scatter graph
plt.title("Depth and diameter of Synechococcus")
plt.xlabel("Diameter (micrometers)")
plt.ylabel("Depth (m)")

plt.ylim(225, 0)

syn_x_axis1 = s1["diameter_synecho"].to_list()
syn_y_axis1 = s1["depth"].to_list()

syn_x_axis4 = s4["diameter_synecho"].to_list()
syn_y_axis4 = s4["depth"].to_list()

plt.scatter(syn_x_axis1, syn_y_axis1, label = "Station 1")
plt.scatter(syn_x_axis4, syn_y_axis4, label = "Station 4")

plt.legend(loc = "upper right")
```

```python
# Prochlorococcus depth and abundance scatter graph
plt.title("Depth and Abundance of Prochlorococcus")
plt.xlabel("Adundance")
plt.ylabel("Depth (m)")

plt.ylim(225, 0)

pro_abun_x_axis1 = s1["cell_abundance_prochloro"].to_list()
pro_abun_y_axis1 = s1["depth"].to_list()

pro_abun_x_axis4 = s4["cell_abundance_prochloro"].to_list()
pro_abun_y_axis4 = s4["depth"].to_list()

plt.scatter(pro_abun_x_axis1, pro_abun_y_axis1, label = "Station 1")
plt.scatter(pro_abun_x_axis4, pro_abun_y_axis4, label = "Station 4")

plt.legend(loc = "lower right")
```

```python
# Synechococcus depth and abundance scatter graph
plt.title("Depth and abundance of Synechococcus")
plt.xlabel("Abundance")
plt.ylabel("Depth (m)")

plt.ylim(225, 0)

s1 = depth_file.loc[depth_file["station"] == 1]
s18 = depth_file.loc[depth_file["station"] == 18]

syn_abun_x_axis1 = s1["cell_abundance_synecho"].to_list()
syn_abun_y_axis1 = s1["depth"].to_list()

syn_abun_x_axis18 = s18["cell_abundance_synecho"].to_list()
syn_abun_y_axis18 = s18["depth"].to_list()

plt.scatter(syn_abun_x_axis1, syn_abun_y_axis1, label = "Station 1")
plt.scatter(syn_abun_x_axis18, syn_abun_y_axis18, label = "Station 18")

plt.legend(loc = "lower right")
```
