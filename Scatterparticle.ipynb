```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Reading in the feather file
g3_stations = pd.read_feather('g3_sample.feather')
```

```python
def particle_scatter_with_depth(feather_file, y_axis, station_num, replicate_letter, depth, y_col_name, yflag = True):
    '''
    Inputs:
    feather_file = string of the feather file name
    y_axis = string of the title for the y axis
    station_num = int of wanted station number
    y_col_name = string of the column name for dependent variable
    yflag = when declared as "True" or not mentioned at all in call, y axis values are put
    in log values; when declared as "False" y axis values are not put in log values

    Returns: scatter graph of syn and picoeuk based on forward scatter and 
    dependent varaible
    '''

    df = pd.read_feather(feather_file)
    
    #Creating plot
    plt.title('Particle Data')
    plt.xlabel('Foward scatter')
    plt.ylabel(y_axis)
    
    syn = df.loc[(df['station'] == station_num) & (df['pop'] == 'synecho') &
                (df['replicate'] == ' ' + replicate_letter) & (df['depth'] == depth)]
    pico = df.loc[(df['station'] == station_num) & (df['pop'] == 'picoeuk') &
                (df['replicate'] == ' ' + replicate_letter) & (df['depth'] == depth)]
    pro = df.loc[(df['station'] == station_num) & (df['pop'] == 'prochloro') &
                (df['replicate'] == ' ' + replicate_letter) & (df['depth'] == depth)]
    
    # synecho data
    x_syn = np.log(syn['norm.scatter'].tolist())
    if yflag:
        y_syn = np.log(syn[y_col_name].tolist())
    else:
        y_syn = syn[y_col_name].tolist()
    
    # picoeukaryote data
    x_pico = np.log(pico['norm.scatter'].tolist())
    if yflag:
        y_pico = np.log(pico[y_col_name].tolist())
    else:
        y_pico = pico[y_col_name].tolist()

    # prochloro data
    x_pro = np.log(pro['norm.scatter'].tolist())
    if yflag:
        y_pro = np.log(pro[y_col_name].tolist())
    else:
        y_pro = pro[y_col_name].tolist()
    
    # plotting synecho particles
    plt.scatter(x_syn, y_syn, s = 15, label = 'Synechococcus', alpha = 0.5, color = 'orange')
    
    # plotting picoeuk particles
    plt.scatter(x_pico, y_pico, s = 15, label = 'Picoeukaryotes', alpha = 0.5, color = 'lightgreen')

    # plotting prochloro particles
    plt.scatter(x_pro, y_pro, s = 15, label = 'Prochlorococcus', alpha = 0.5, color = 'dodgerBlue')
    
    plt.legend(loc = 'best')
```

```python
def particle_scatter(feather_file, y_axis, station_num, replicate_letter, y_col_name, yflag = True):
    '''
    Inputs:
    feather_file = string of the feather file name
    y_axis = string of the title for the y axis
    station_num = int of wanted station number
    y_col_name = string of the column name for dependent variable
    yflag = when declared as "True" or not mentioned at all in call, y axis values are put
    in log values; when declared as "False" y axis values are not put in log values

    Returns: scatter graph of syn and picoeuk based on forward scatter and 
    dependent varaible
    '''

    df = pd.read_feather(feather_file)
    
    #Creating plot
    plt.title('Particle Data')
    plt.xlabel('Foward scatter')
    plt.ylabel(y_axis)
    
    syn = df.loc[(df['station'] == station_num) & (df['pop'] == 'synecho') &
                (df['replicate'] == ' ' + replicate_letter)]
    pico = df.loc[(df['station'] == station_num) & (df['pop'] == 'picoeuk') &
                (df['replicate'] == ' ' + replicate_letter)]
    pro = df.loc[(df['station'] == station_num) & (df['pop'] == 'prochloro') &
                (df['replicate'] == ' ' + replicate_letter)]
    
    # synecho data
    x_syn = np.log(syn['norm.scatter'].tolist())
    if yflag:
        y_syn = np.log(syn[y_col_name].tolist())
    else:
        y_syn = syn[y_col_name].tolist()
    
    # picoeukaryote data
    x_pico = np.log(pico['norm.scatter'].tolist())
    if yflag:
        y_pico = np.log(pico[y_col_name].tolist())
    else:
        y_pico = pico[y_col_name].tolist()

    # prochloro data
    x_pro = np.log(pro['norm.scatter'].tolist())
    if yflag:
        y_pro = np.log(pro[y_col_name].tolist())
    else:
        y_pro = pro[y_col_name].tolist()
    
    # plotting synecho particles
    plt.scatter(x_syn, y_syn, s = 15, label = 'Synechococcus', alpha = 0.5, color = 'orange')
    
    # plotting picoeuk particles
    plt.scatter(x_pico, y_pico, s = 15, label = 'Picoeukaryotes', alpha = 0.5, color = 'lightgreen')

    # plotting prochloro particles
    plt.scatter(x_pro, y_pro, s = 15, label = 'Prochlorococcus', alpha = 0.5, color = 'dodgerBlue')
    
    plt.legend(loc = 'best')
```
